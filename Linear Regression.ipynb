```python
 # Import required libraries
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
from sklearn.linear_model import LinearRegression
```

```python
# Define the path to the CSV file
file_path = Path("Resources/housing_data.csv")

# Read the CSV file into a DataFrame
df_housing = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df_housing.head()
```

```python
df_housing.info()
```

```python
# Example: Basic data exploration
df_housing.describe()
```

```python
#Scatter: Plotting with hvPlot
housing_plot = df_housing.hvplot.scatter(
    x="sqft_living",
    y="price",
    title="Expected Price Based on Square Feet"
)
housing_plot
```

```python
#Bar: Plotting with hvPlot
plot = df_housing.hvplot.bar(x='sqft_living', y='price', title='Price per square foot')
hvplot.show(plot)
```

```python
#Line: Plotting with hvPlot
plot = df_housing.hvplot.line(x='sqft_living', y='price', title='Price per square foot')
hvplot.show(plot)
```

```python
# Create the X set by using the `reshape` function to format the housing data as a single column array.
X = df_housing["sqft_living"].values.reshape(-1, 1)

# Display sample data
X[:5]
```

```python
# Create an array for the dependent variable y with the housing data
y = df_housing["price"]
```

```python
 # Create a model with scikit-learn
model = LinearRegression()
```

```python
# Fit the data into the model
model.fit(X, y)
```

```python
# Display the slope
print(f"Model's slope: {model.coef_}")
```

```python
# Display the y-intercept
print(f"Model's y-intercept: {model.intercept_}")
```

```python
# Display the model's best fit line formula
print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]}X")
```

```python
 # Make predictions using the X set
predicted_y_values = model.predict(X)
```

```python
 # Create a copy of the original data
df_housing_predicted = df_housing.copy()

# Add a column with the predicted sales values
df_housing_predicted["housing_predicted"] = predicted_y_values

# Display sample data
df_housing_predicted.head()
```

```python
# Create a line plot of sqft_living versus the predicted housing values
best_fit_line = df_housing_predicted.hvplot.line(
    x = "sqft_living",
    y = "housing_predicted",
    color = "red"
)
best_fit_line
```

```python
# Superpose the original data and the best fit line
housing_plot * best_fit_line
```

```python
#Linear Regression Model Assessment

 # Import relevant metrics from scikit-learn
from sklearn.metrics import mean_squared_error, r2_score
```

```python
 # Compute metrics for the linear regression model: score, r2, mse, rmse, std
score = model.score(X, y, sample_weight=None)
r2 = r2_score(y, predicted_y_values)
mse = mean_squared_error(y, predicted_y_values)
rmse = np.sqrt(mse)
std = np.std(y)

# Print relevant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is {std}.")
```
