```python
# Initial imports
import pandas as pd
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
 # Loading data
file_path = Path("Resources/housing_data.csv")
df_housing = pd.read_csv(file_path)
df_housing.head()
```

```python
# Heatmap
# Drop non numeric fields 
data_df = df_housing.select_dtypes(include=[float, int])

# Generate correlation matrix
correlation_matrix = data_df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='cividis')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter Plot of Living Area vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=df_housing)
plt.title('Living Area vs. Price')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Price')
plt.show()
```

```python
#Plot of Houses by Years Built

plt.figure(figsize=(12, 6))
sns.histplot(df_housing['yr_built'], bins=20, color='red', kde=True)
plt.xlabel('Year Built')
plt.ylabel('Number')
plt.title('Houses Built by Year')
plt.show()
```

```python
#Plot of Property Quality

plt.figure(figsize=(12, 6))
sns.countplot(x='condition', data=df_housing)
plt.xlabel('Condition')
plt.ylabel('Number')
plt.title('NUmber of House by Condition')
plt.show()
```

```python
# Plot of Properties by Zipcode
plt.figure(figsize=(12, 6))
sns.countplot(x='zipcode', data=df_housing)
plt.xlabel('Zipcode')
plt.ylabel('Number')
plt.title('Number of Properties by Zipcode')
plt.xticks(rotation=90)
plt.show()
```

```python
#Plot of Bedrooms by Price

plt.figure(figsize=(12, 6))
sns.boxplot(x='bedrooms', y='price', data=df_housing)
plt.title('Number of Bedrooms by Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```
